# Making our own Gradient Descent Class

```python
class GDRegressor:

    def __init__(self, epochs, learning_rate):
        self.m = 100
        self.b = 100
        self.epochs = epochs
        self.lr = learning_rate

    def fit(self, X, y):
        for i in range(self.epochs):
            self.b = self.b - self.lr * (-2) * np.sum(y - self.m * X.ravel() - self.b)
            self.m = self.m - self.lr * (-2) * np.sum ((y - self.m * X.ravel() - self.b) * X.ravel())

        print(self.m, self.b)

    def predict(self, X):
        return self.m * X + self.b
```

```python
gd = GDRegressor(200, 0.01)
```

```python
gd.fit(X,y)
```
