```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/cement-manufacturing-concrete-dataset/concrete.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
X = df.drop(['strength'], axis=1)
y = df['strength']
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print(np.mean(cross_val_score(lr, X, y, cv=10)))
```

### without binning, getting accuracy of 57 %

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
kbins = KBinsDiscretizer(n_bins = 10, encode='ordinal', strategy='quantile')
```

```python
X_train_bin = kbins.fit_transform(X_train)
```

```python
X_test_bin = kbins.transform(X_test)
```

```python
kbins.bin_edges_[0]
```

```python
X_train
```

```python
X_train_bin.shape
```

```python
output = pd.DataFrame(X_train_bin, columns = X_train.columns)
```

```python
output
```

```python
i = pd.concat( [output.reset_index(drop=True), X_train.reset_index(drop=True)], axis=1)
```

```python
i
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train_bin, y_train)
```

```python
lr.score(X_test_bin, y_test)
```
