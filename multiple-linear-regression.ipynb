# Multiple Linear Regression Class creation

```python
class MeraLR:
    
    def __init__(self):
        self.m = None
        self.b = None
    def fit(self, X_train, y_train):
        X_final = np.insert(X_train, 0, 1, axis=1)

        # Coefficients 
        betas = np.linalg.inv(np.dot(X_final.T, X_final)).dot(X_final.T).dot(y_train)
        self.b = betas[0]
        self.m = betas[1:]

        print(self.m)
        print(self.b)

    def predict(self, X_test):
        y_pred =  np.dot(X_test, self.m) + self.b
        return y_pred
```

```python
reg = MeraLR()
```

```python
reg.fit(X_train, y_train)
```

```python
reg.b
```

```python
y_pred = reg.predict(X_test)
```

```python
print(r2_score(y_test, y_pred))
```
